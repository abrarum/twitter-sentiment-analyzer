```python
import numpy as np
import pandas as pd
import re
#pd.set_option('display.min_rows', 100)

#master data
master_arr = []
train_data = []
test_data = []

def readtxt():
    print('reading txt file and converting to pandas dataframe ...')
    ###
    global master_arr
    
    master_arr = pd.read_csv('./data.txt', delimiter="\t", names=["id", "tag", "score", "tweet"])

def savetocsv():
    print('saving the dataframe into csv ...')
    ###
    global master_arr
    
    master_arr.to_csv('./datatable.csv')

def dataset():
    print("distributing data into 80% training and 20% testing data ...")
    ###
    global train_data
    global test_data
    
    total_rows = int(master_arr.shape[0])
    train_data = master_arr[:int(total_rows*0.8)]
    test_data = master_arr[int(total_rows*0.8):]
    
def casefolding():
    print("preprocessing data ...")
    ###
    print("casefolding initiating ...")
    
    global train_data
    global test_data
    
    myset = [train_data, test_data]
    for i in myset:
        
        print("folding the cases to lowercase ...")
        i.loc[:, 'tweet'] = [tweet.lower() for tweet in i['tweet']]
        
        print("replacing emojis with texts ...")
        
        #i.loc[:, 'tweet'] = [emoji.demojize(tweet) for tweet in i['tweet']]
        
        print("removing links ...")
        i.loc[:, 'tweet'] = [re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', tweet) for tweet in i['tweet']]
        
        print("removing usernames ...")
        i.loc[:, 'tweet'] = [re.sub('@[^\s]+', 'USER', tweet) for tweet in i['tweet']]
        
        print("removing hash from hashtags ...")
        i.loc[:, 'tweet'] = [re.sub(r'#([^\s]+)', r'\1', tweet) for tweet in i['tweet']]
        
        print("removing unnecessary characters ...")
        i.loc[:, 'tweet'] = [re.sub(r"[\[\]\(\){},.!\?;:\-_'\"]", '' , tweet) for tweet in i['tweet']]
        
        print("removing stop words")
        
    print(train_data['tweet'])
    
if __name__ == '__main__':
    readtxt()
    savetocsv()
    dataset()
    
    #preprocessing
    cased = casefolding()
```
